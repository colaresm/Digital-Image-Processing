```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/img.png",cv2.IMREAD_GRAYSCALE)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB,)
image
```

```python
def add_gaussian_noise(image, mean=0, std=25):
    noise = np.random.normal(mean, std, image.shape)
    noisy_image = image + noise
    noisy_image = np.clip(noisy_image, 0, 255)
    return noisy_image.astype(np.uint8)

def apply_mean_filter(image, kernel_size=3):
    filtered_image = cv2.blur(image, (kernel_size, kernel_size))
    return filtered_image

def apply_median_filter(image, kernel_size=3):
    filtered_image = cv2.medianBlur(image, kernel_size)
    return filtered_image


mean = 0
std = 20
noisy_image = add_gaussian_noise(image, mean, std)
filtered_image = apply_mean_filter(noisy_image, kernel_size=5)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Imagem Original")
plt.imshow(image)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Imagem com Rudo Gaussiano")
plt.imshow(noisy_image)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Imagem com Filtro da Mdia")
plt.imshow(filtered_image)
plt.axis("off")

plt.show()
```

```python
filtered_image = apply_median_filter(noisy_image, kernel_size=3)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Imagem Original")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Imagem com Rudo Gaussiano")
plt.imshow(noisy_image, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Imagem com Filtro da Mediana")
plt.imshow(filtered_image, cmap="gray")
plt.axis("off")

plt.show()
```
